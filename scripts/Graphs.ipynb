# Iterative-solver Graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv', delimiter=',')
df = df.sort_values(by=['method'])
df = df.sort_values(by=['matrix_name', 'method'])
```

```python
#tol = np.array([1.000000e-03, 1.000000e-05, 1.000000e-07, 1.000000e-09])
tol = np.array([9, 7, 5, 3])
grad_rel_err = df['relative_error'][0:16]
grad_con_rel_err = df['relative_error'][16:32]
jac_rel_err = df['relative_error'][32:48]
gauss_rel_err = df['relative_error'][48:64]
```

## tolerance/relative_error graphs, difference between the 4 methods

```python
plt.plot(grad_rel_err[0:4], tol, label='Gradiente')
plt.plot(grad_con_rel_err[0:4], tol, label='Gradiente coniugato')
plt.plot(jac_rel_err[0:4], tol, label='Jacobi')
plt.plot(gauss_rel_err[0:4], tol, label='Gauss-Seidel')

plt.scatter(grad_rel_err[0:4], tol)
plt.scatter(grad_con_rel_err[0:4], tol)
plt.scatter(jac_rel_err[0:4], tol)
plt.scatter(gauss_rel_err[0:4], tol)

plt.legend()

plt.ylabel('Tolleranza')
plt.xlabel('Errore relativo')
plt.title('Spa1 tolleranza/errore_relativo')
#plt.axis([0.001, 0.02, 0.0000, 0.0012])

plt.show()
```
