# Iterative-solver Graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import log
```

```python
def plot_matrix_data(df, key1, key2, value1, value2, plt_label, xName, yName, xLabel=None, yLabel=None):
    plt.xlabel(xLabel if xLabel is not None else xName)
    plt.ylabel(yLabel if yLabel is not None else yName)
    plt.plot(df.loc[(df[key1]==value1) & (df[key2]==value2)][xName], df.loc[(df[key1]==value1) & (df[key2]==value2)][yName])
    plt.scatter(df.loc[(df[key1]==value1) & (df[key2]==value2)][xName], df.loc[(df[key1]==value1) & (df[key2]==value2)][yName], label=plt_label)
    
def multi_plots(df, key1, key2, title, xName, yName, xLabel, yLabel):
    for value1 in df[key1].unique():
        for value2 in df[key2].unique():
            plt.xscale('log')
            plt.yscale('log')
            plot_matrix_data(df, key1, key2, value1, value2, value2, xName, yName, xLabel, yLabel)
        plt.title(title + ' - ' + value1)
        plt.legend()
        plt.show()
```

## Tolerance vs Relative error graphs, difference between the 4 methods

```python
multi_plots(df, 'matrix_name', 'method', 'Tolerance vs Relative error', 'tolerance', 'relative_error', 'Tolerance', 'Relative error')
```

## Tolerance vs Elapsed time graphs, difference between the 4 methods

```python
multi_plots(df, 'matrix_name', 'method', 'Tolerance vs Elapsed time', 'tolerance', 'elapsed_ms', 'Tolerance', 'Elapsed time (ms)')
```
